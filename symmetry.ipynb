```python
#Téléchargement du fichier CIF
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
with MPRester("AzZfK75KWoi3UDJbKM3u") as m:
        structure = m.get_structure_by_material_id("mp-568588")
        w = CifWriter(structure)
        w.write_file('NdTlSe2_mp-568588_symmetrized.cif')
```

```python
#Première symétrie
print('Première symétrie : plan miroir défini par (-x+y,y,z) et appliquée sur un atome Tl de coordionnées (0.67,0.33,0.83).')
from pymatgen.core.operations import SymmOp

symetrie1 = SymmOp.from_xyz_string('-x+y,y,z')
print('La matrice de rotation et le vecteur de translation associés à cette symétrie sont les suivants :')
print(symetrie1)

r1 = symetrie1.operate((0.67,0.33,0.83))
print('Les coordonnées de cet atome après application de la symétrie sont les suivantes :')
print(r1)
```

```python
#Deuxième symétrie
print('Deuxième symétrie : rotation d ordre 3 définie par (-y,x-y,z) et appliquée sur un atome Nd de coordonnées (1,0,0).')
from pymatgen.core.operations import SymmOp

symetrie2 = SymmOp.from_xyz_string('-y,x-y,z')
print('La matrice de rotation et le vecteur de translation associés à cette symétrie sont les suivants :')
print(symetrie2)

r2 = symetrie2.operate((1.0, 0.0, 0.0))
print('Les coordonnées de cet atome après application de la symétrie sont les suivantes :')
print(r2)
```
