```python
#Téléchargement du fichier CIF
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.core.operations import SymmOp
from IPython.display import Image
from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))

with MPRester("AzZfK75KWoi3UDJbKM3u") as m:
        structure = m.get_structure_by_material_id("mp-568588")
```

```python
printmd("**Première symétrie**")
print("Plan miroir défini par (-x+y,y,z) et appliqué sur un atome Tl de coordionnées (0.67,0.33,0.83).")

symetrie1 = SymmOp.from_xyz_string('-x+y,y,z')
print("La matrice de rotation et le vecteur de translation associés à cette symétrie sont les suivants :")
print(symetrie1)

r1 = symetrie1.operate((0.67,0.33,0.83))
print("Les coordonnées de cet atome après application de la symétrie sont les suivantes :")
print(r1)

print("Cette première symétrie est illustrée sur les deux images suivantes :")
Image(filename='Symetrie1.PNG')
```

```python
printmd("**Deuxième symétrie**")
print("Rotation d'ordre 3 définie par (-y,x-y,z) et appliquée sur un atome Nd de coordonnées (1,0,0).")

symetrie2 = SymmOp.from_xyz_string('-y,x-y,z')
print("La matrice de rotation et le vecteur de translation associés à cette symétrie sont les suivants :")
print(symetrie2)

r2 = symetrie2.operate((1.0, 0.0, 0.0))
print("Les coordonnées de cet atome après application de la symétrie sont les suivantes :")
print(r2)

print("A nouveau, deux images permettent d'illustrer cette symétrie.")
Image(filename='Symetrie2.PNG')
```

```python
printmd("**Troisième symétrie**")
print("Rotation hélicoïdale n_1/3 définie par (-y-1/3,x-y-1/3,z+1/3) et appliquée sur un atome Se de coordonnées (1,0,0.74).")

symetrie3 = SymmOp.from_xyz_string('-y-1/3,x-y-1/3,z+1/3')
print("La matrice de rotation et le vecteur de translation associés à cette symétrie sont les suivants :")
print(symetrie3)

r3 = symetrie3.operate((1.0,0.0,0.74))
print("Les coordonnées de cet atome après application de la symétrie sont les suivantes :")
print(r3)

print("Cette troisième symétrie est donc une rotation hélicoïdale n_1/3. C'est-à-dire qu'il s'agit d'une rotation de 2*pi/3 suivie ou précédée d'une translation de 1/3. Cette symétrie est appliquée sur un atome Se de coordonnées (1,0,0.74) et à nouveau 2 images illustrent la symétrie :")
Image(filename='Symetrie3.PNG')
```
