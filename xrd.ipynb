```python
#Téléchargement du fichier CIF
from pymatgen.ext.matproj import MPRester
from matplotlib import pyplot as plt
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import XRDCalculator
with MPRester("AzZfK75KWoi3UDJbKM3u") as m:
        structure = m.get_structure_by_material_id("mp-568588")
        w = CifWriter(structure)
        w.write_file('NdTlSe2_mp-568588_symmetrized.cif')
#Obtention des pics du diffractogramme pour un rayonnement Cu de longueur d'onde 1.54060 Angström
A = XRDCalculator(1.54060)
B = A.get_pattern(structure)
C = A.get_plot(structure, two_theta_range=(0, 90), annotate_peaks=True, ax=None, with_labels=True, fontsize=8)
print(C)
#print(B)
print('Le premier pic a une intensité 1.28300512e-01 pour un angle de diffraction 2*theta égal à 11.36879459. Les indices de Miller associés à ce pic sont (1 1 1).')
print('Le deuxième pic a une intensité 9.32653189e+00 pour un angle de diffraction 2*theta égal à 22.85146163. Les indices de Miller associés à ce pic sont (2 2 2)')
print('Le troisième pic a une intensité 2.73239057e+00 pour un angle de diffraction 2*theta égal à 24.00970902. Les indices de Miller associés à ce pic sont (1 0 0)')
```
